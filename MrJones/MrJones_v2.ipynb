```python
# imports two different libraries: matplotlib to do the plotting and numpy to do the math and generate arrays
import matplotlib.pyplot as plt
import numpy as np
# Defines (names) two contstants (objects): G as the gravational constant and two different masses
G = 6.674e-11   # Gravational constant
m1 = 1000       # Mass 1 in kg
m2 = 500        # Mass 2 in kg
# Creates (names) a distance array of values from 01 to 10 with 100 evenly spaced values (to smooth out the curve)
distances = np.linspace(1, 10, 100)
# Calculates the gravational force between to objects of specified mass
forces = G * (m1 * m2) / (distances**2)
# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(distances, forces, 'b-', label=f'Masses: {m1}kg and {m2}kg')
#labels, title, grid, legend
plt.xlabel('Distances (m)')
plt.ylabel('Gravational Force (N)')
plt.title('Gravational Force vs Distance')
plt.grid(True, alpha=0.3)
plt.legend()
# shows the plot, graph and all customizations
plt.show()
```
