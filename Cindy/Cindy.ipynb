```python
import matplotlib.pyplot as plt
import numpy as np

G = 6.674e-11   
m1 = 1000       
m2 = 500       

distances = np.linspace(1, 10, 100)

forces = G * (m1 * m2) / (distances**2)

plt.figure(figsize=(10, 6))
plt.plot(distances, forces, 'b-', label=f'Masses: {m1}kg and {m2}kg')

plt.xlabel('Distances (m)')
plt.ylabel('Gravational Force (N)')
plt.title('Gravational Force vs Distance')
plt.grid(True, alpha=0.3)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = [0, 1, 2, 3, 4, 5]
linear_y = [0, 1, 2, 3, 4, 5]     
quadratic_y = [0, 1, 4, 9, 16, 25] 

plt.plot(x_values, linear_y, 'b-o' ,label='Linear (y = x)') 
plt.plot(x_values, quadratic_y, 'r-s', label='Quadtratic (y = x^2)') 

plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Comparison of Linear and Quadratic Growth')
plt.grid(True, alpha=0.3)
plt.legend

plt.xlim(-0.5,5.3)
plt.ylim(-1, 26)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

G = 6.674e-11   
m1 = 1000       
m2 = 500  
m3 = 2000 
m4 = 3000 


distances = np.linspace(1, 10, 100)

forces1 = G * (m1 * m2) / (distances**2)
forces2 = G * (m3 * m2) / (distances**2)
forces3 = G * (m4 * m2) / (distances**2)

plt.figure(figsize=(10, 6))
plt.plot(distances, forces1, 'b-', label=f'Masses: {m1}kg and {m2}kg')
plt.plot(distances, forces2, 'g-', label=f'Masses: {m3}kg and {m2}kg')
plt.plot(distances, forces3, 'r-', label=f'Masses: {m4}kg and {m2}kg')

plt.xlabel('Distances (m)')
plt.ylabel('Gravational Force (N)')
plt.title('Gravational Force vs Distance')
plt.grid(True, alpha=0.3)
plt.legend()
plt.yscale('log')

plt.show()
```
