```python
import matplotlib.pyplot as plt
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.plot(x, y)
plt.title("Quadratic Relationship")
plt.xlabel("X value")
plt.ylabel("Y value (x^2)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
x_values = [0, 1, 2, 3, 4, 5]
linear_y = [0, 1, 2, 3, 4, 5]
quadratic_y = [0, 1, 4, 9, 16, 25]
plt.plot(x_values, linear_y, 'b-o', label='Linear (y=x)')
plt.plot(x_values, quadratic_y,'r-s' ,label='Quadratic(y=x^2)')
plt.xlabel('X Values')
plt.ylabel('Y Values (X²)')
plt.title('Quadratic Relationship')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
G = 6.674e-11
m1 = 1000
m2 = 500
distance = np.linspace(1, 10, 100)
forece = G * (m1*m2)/(distance**2)
plt.figure(figsize=(10, 6))
plt.plot(distance, forece, 'b-', label=f'Masses: {m1}kg and {m2}kg')
plt.xlabel('Distances(m)')
plt.ylabel('Gravational Force(N)')
plt.title('Gravational Forece vs Distance')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
