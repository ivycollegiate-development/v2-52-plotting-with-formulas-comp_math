```python
import matplotlib.pyplot as plt
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.plot(x, y)
plt.title("Quadratic Relationship")
plt.xlabel("X value")
plt.ylabel("Y value (x^2)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
x_values = [0, 1, 2, 3, 4, 5]
linear_y = [0, 1, 2, 3, 4, 5]
quadratic_y = [0, 1, 4, 9, 16, 25]
plt.plot(x_values, linear_y, 'b-o', label='Linear (y=x)')
plt.plot(x_values, quadratic_y,'r-s' ,label='Quadratic(y=x^2)')
plt.xlabel('X Values')
plt.ylabel('Y Values (X)')
plt.title('Quadratic Relationship')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
